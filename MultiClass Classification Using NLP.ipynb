# Importing Libraries And Loading NLP(Binary Text Classification) Dataset From Kaggle.com

```python
import numpy as np
import pandas as pd
df = pd.read_csv("tweets.csv")
df.info()
df
```

# Selecting Only Concerned Columns

```python
df=df[['hashtags','like_count','rt_count']]
df
```

```python
#Combining words in a single phrase
def unlisting(*x):
    g=''
    a=str(list(x))
    b=a.replace('[','')
    c=b.replace(']','')
    d=c.replace('"','')
    e=d.replace('\'','')
    f=e.split(',')
    for i in f:
        g+=str(i)
       
    return g
df['hashtags1']=df['hashtags'].apply(unlisting)
df
```

# Applying Naive Bayes Algorithm first on column like_count

```python
df1=df[['hashtags1','like_count']]
df1
```

```python
#Sampling 20%
df1.like_count.value_counts() *20/100
```

```python
#Sampling 80%
df1.like_count.value_counts() *80/100
```

```python
#0-Negative
a=int(150939.2/50)
df1[df1.like_count==0][:a]
```

```python
#1-Somewhat Negative
b=int(74715.2/50)
df1[df1.like_count==1][:b]
```

```python
#2-Neutral
c=int(38392.0/50)
df1[df1.like_count==2][:c]
```

```python
#3-Somewhat Positive
d=int(22449.6/50)
df1[df1.like_count==3][:d]
```

```python
#4-Positive
e=int(14607.2/50)
df1[df1.like_count==4][:e]
```

```python
#testing data
test = pd.concat([df1[df1.like_count==0][:a],df1[df1.like_count==1][:b],df1[df1.like_count==2][:c],
                  df1[df1.like_count==3][:d],df1[df1.like_count==4][:e]])[["hashtags1","like_count"]]
print(len(test))
display(test)
```

```python
#training data        
a=int(37735.2/50)
b=int(18678.8/50)
c=int(9598.0/50)
d=int(5612.4/50)
e=int(3651.8/50)
p=int(150939.2/20)
q=int(74715.2/20)
r=int(38392.0/20)
s=int(22449.6/20)
t=int(14607.2/20)
v=df1[df1.like_count==0][a:p]
w=df1[df1.like_count==1][b:q]
x=df1[df1.like_count==2][c:r]
y=df1[df1.like_count==3][d:s]
z=df1[df1.like_count==4][e:t]
train = pd.concat([df1[df1.like_count==0][a:p],df1[df1.like_count==1][b:q],df1[df1.like_count==2][c:r],
                    df1[df1.like_count==3][d:s],df1[df1.like_count==4][e:t]])[["hashtags1","like_count"]]
print(len(train))
display(train)
```

```python
X=train[["hashtags1"]] #corpus all text from your dataset
X
```

```python
#Convert corpus into vector with sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
#list of text document
courpus = X.hashtags1.values
# create transorm
vectorizer = TfidfVectorizer()
# tokeniz and build vocab
vectorizer.fit(courpus)
#summarize
print(vectorizer.vocabulary_)
```

```python
#Transform Training and Testing input text into vector using above
train_x = vectorizer.transform(train.hashtags1.values).toarray()
train_y = train.like_count.values
test_x = vectorizer.transform(test.hashtags1.values).toarray()
test_y = test.like_count.values
display("Train_x",train_x)
display("Train_y",train_y)
display("Test_x",test_x)
display("Test_y",test_y)
print(len(train_x),len(train_y))
print(len(test_x),len(test_y))
```

```python
#Bernoulli Naive Bayes for Binary Classification
from sklearn.naive_bayes import BernoulliNB
clf = BernoulliNB()
clf.fit(train_x,train_y)
```

```python
#Now Apply prediction on Unseendata
y_predict = clf.predict(test_x)
y_predict
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

cm = confusion_matrix(test_y,y_predict)
cm
```

```python
accuracy_score(test_y, y_predict)
```

```python
#sklearn.metrics.classification_report
from sklearn.metrics import classification_report
#                           Actual Unseen data
print(classification_report(test_y,y_predict))
```
